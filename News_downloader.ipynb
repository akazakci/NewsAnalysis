# Suivi et analyse automatisé des articles sur le web

Nous allons créer un environnement de travail, charger quelques outils en mémoire et télécharger les derniers articles qui ont apparu sur un site web donné. Nous avons choisi lemonde.fr à titre d'illustration, mais on peut modifier cette source comme on veut. Pour le travail envisagé, nous pouvons créer une liste de sources à scanner et appliquer les étapes de téléchargements suivantes à toutes ses resources par intervalle régulier.

```python
import newspaper
lemonde_paper = newspaper.build('http://lemonde.fr', memoize_articles=False)
```

```python
len(lemonde_paper.articles) 
```

```python
for article in lemonde_paper.articles:
    print (article.url) 
```

# Nous allons télécharger ces articles sous format html (cela va durer qqs minutes)

```python
for article in lemonde_paper.articles:
    article.download()
```
